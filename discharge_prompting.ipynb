```python
import tiktoken

def count_tokens(text):
    # Load the tokenizer for the specified model
    encoding = tiktoken.encoding_for_model('gpt-4o')

    # Encode the text to get tokens
    tokens = encoding.encode(text)

    # Return the number of tokens
    return len(tokens)

# Example usage
token_count = count_tokens(raw_summary)
print(f"Number of tokens: {token_count}")
```

```python
individual_prompt = '''
You are a medical AI that extracts key concepts and phrases from patient discharge summaries.
Your task is to take a labeled section of a discharge summary and extract the main medical concepts
that would be key to a doctor's understanding of the case. Output a comma-separated list of
about 10 terms or phrases. Medical Jargon is okay; do not explain any of the terms.
All extracted terms/phrases must be derived from my input -- do not infer outside terms.

Certain identifying patient data may be redacted and replaced with underscores like "___" or may be too short to analyze.
If a section I provide you is completely redacted or too short to contain meaningful information, simply output 'N/A'
My next input will contain the data for you to process.
'''
```

```python
check_prompt = '''
You are a medical AI that ensures another AI has not hallucinated when condensing sections of patient discharge summaries.
Your task is to compare a comma-separated list of key terms and phrases, generated by AI, to a longer text that was used to create the aforementioned comma-separated list.
Hallucinations for our purposes are encountered when the comma-separated list contains terms or phrases unrelated to the longer text.

If you do not detect any hallucinations, simply output "Pass" and nothing else. If you do detect a hallucination, output "Fail", and on a separate line, a brief explanation of why you think it is a hallunication.

I will provide two examples below, one of a passing input, and one of a failing input.

**BEGIN PASSING EXAMPLE**

HUMAN-WRITTEN LONGER TEXT: Family History:\nShe a total of five siblings, but she is not  \ntalking to most of them. She only has one brother that she is in \n \ntouch with and lives in ___. She is not aware of any  \nknown GI or liver disease in her family.  \nHer last alcohol consumption was one drink two months ago. No  \nregular alcohol consumption. Last drug use ___ years ago. She  \nquit smoking a couple of years ago.

AI-GENERATED KEY TERMS: five siblings, not talking to most siblings, one brother in contact, no known GI disease, no known liver disease, last alcohol consumption two months ago, no regular alcohol consumption, last drug use ___ years ago, quit smoking, couple of years ago

YOUR OUTPUT: Pass

**END PASSING EXAMPLE**

**BEGIN FAILING EXAMPLE**

HUMAN-WRITTEN LONGER TEXT: Family History:\nShe a total of five siblings, but she is not  \ntalking to most of them. She only has one brother that she is in \n \ntouch with and lives in ___. She is not aware of any  \nknown GI or liver disease in her family.  \nHer last alcohol consumption was one drink two months ago. No  \nregular alcohol consumption. Last drug use ___ years ago. She  \nquit smoking a couple of years ago.

AI-GENERATED KEY TERMS: seven siblings, not talking to most siblings, one brother in contact, no known GI disease, no known liver disease, last alcohol consumption last week, no regular alcohol consumption, last drug use ___ years ago, quit smoking, couple of years ago, family history of schizophrenia

YOUR OUTPUT: Fail. \nThe AI output says the patient has seven siblings when they really have five, incorrectly asserts their last alcohol consumption as last week instead of two months ago, and incorrectly
states that the patient has a history of schizophrenia.

**END FAILING EXAMPLE**

Begin processing after my next message which will follow the above examples.

'''
```

```python
detected_hallucinations = []
```

```python
def detect_hallucination(row_num, label, raw_section, condensed_section):
    system_message = {"role": "system", "content": check_prompt}
    user_prompt = f'''
    HUMAN-WRITTEN LONGER TEXT: {raw_section}

    AI-GENERATED KEY TERMS: {condensed_section}
    '''
    user_message = {"role": "user", "content": user_prompt}
    chat_history = [system_message, user_message]

    orig_output = send_message(chat_history, 75)
    if orig_output == 'Pass':
        return False, None
    else:

        correction_prompt = '''
        Okay, now regenerate the AI-generated key terms, but remove the terms you identified as hallucinations. Keep the rest of the AI-generated key terms. Simply output the comma-separated list and nothing else.
        '''
        chat_history.append(correction_prompt)
        output = send_message(chat_history, count_tokens(condensed_section))
        detected_hallucinations.append((row_num, label, user_prompt, orig_output, output))

        return True, output

def condense_discharge_summary(row_num):
    system_message = {"role": "system", "content": prompt}
    raw_summary = df.loc[row_num, 'TEXT']
    user_message = {"role": "user", "content": raw_summary}
    num_tokens = count_tokens(raw_summary)

    print('Adding labels...')
    labeled_summmary = send_message([system_message, user_message], num_tokens + 500)
    pieces = labeled_summmary.split('SECTION SEPARATOR')

    condense_discharge_summary_sections(row_num, pieces)

def condense_discharge_summary_sections(row_num, pieces):
    system_message = {"role": "system", "content": individual_prompt}
    i = 0

    for (label, raw_section) in zip(sections, pieces):
        print(f'Processing label: {label}')
        user_message = {"role": "user", "content": raw_section}
        num_tokens = count_tokens(raw_section)
        condensed_section = send_message([system_message, user_message], num_tokens)

        df.loc[row_num, label] = condensed_section

        if condensed_section != 'N/A':
            hallucination_detected, new_condensed_section = detect_hallucination(row_num, label, raw_section, condensed_section)
            if hallucination_detected:
                df.loc[row_num, label] = new_condensed_section
```
